# Data Cleaning

```python
df.isnull().sum()
```

```python
df.drop(columns = ['id', 'backdrop_path', 'homepage', 'imdb_id', 'overview', 
                  'poster_path', 'tagline'], inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Dropping rows that contain empty values

df.dropna(how = 'any', subset = ['genres', 'spoken_languages'], inplace = True)
```

```python
df['production_companies'] = df['production_companies'].fillna(0)
df['production_countries'] = df['production_countries'].fillna(0)
df['release_date'] = df['release_date'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['vote_average'] = df['vote_average'].round(2)
df['popularity'] = df['popularity'].round(2)
```

```python
df[['vote_average', 'popularity']]
```

```python
# Creating a profit column that consists of revenue minnus budget
```

```python
df['profit'] = df['revenue'] - df['budget']
```

```python
df.head()
```

```python
# Creating a return on investment column, which is profit against budget
```

```python
df['return_on_investment'] = df['profit']/df['budget']
```

```python
df.head()
```

```python
df['return_on_investment'] = df['return_on_investment'].round(2)
```

```python
# Creating a release_year and release_month column from release_date
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df['release_year'] = df['release_date'].dt.year
df['release_month'] = df['release_date'].dt.month
```

```python
df.head()
```

```python
# Renaming column
```

```python
df = df.rename(columns = {'vote_average':'rating'})
```

```python
# Splitting the genres column
```

```python
df['genres'].value_counts()
```

```python
df['genres'] = df['genres'].str.split(',').str[0]
df['genres']
```

```python
df['genres'].value_counts()
```

```python
# Splitting the production companies column
```

```python
df['production_companies']
```

```python
df['production_companies'] = df['production_companies'].str.split(',').str[0]
df['production_companies']
```

```python
df['production_companies'].value_counts()
```
